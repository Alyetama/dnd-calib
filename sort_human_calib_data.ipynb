````python
import os
import shutil
import time
from glob import glob
from pathlib import Path

import pandas as pd
import torch
from tqdm import tqdm
````

````python
def change_file_time(file, t):
    timestamp_str = str(t)
    parsed_time = time.strptime(timestamp_str, '%Y-%m-%d %H:%M:%S')
    ts = time.mktime(parsed_time)
    os.utime(file, (ts, ts))
````

````python
df = pd.read_csv('wildlife-insights_4fc92e40-94fe-48d3-a51f-5ac1a03f5e03_project-2007160_data/images_2007160.csv')
len(df)
````

````python
len(df['deployment_id'].unique())
````

````python
df['timestamp'] = pd.to_datetime(df['timestamp'])
````

````python
pickup_date = '2024-03-23'
df = df[df['timestamp'].dt.date == pd.to_datetime(pickup_date).date()]
````

````python
df = df[df['common_name'].isin(['Human-Camera Trapper', 'Human-Faces'])]
len(df)
````

````python
df = df.sort_values(by='timestamp')
````

````python
df.reset_index(drop=True, inplace=True)
````

````python
df['deployment_id'] = df['deployment_id'].str.replace('/', '-')
````

````python
df = df[df['deployment_id'].isin([...])]
````

````python
len(df)
````

````python
df['location'].to_csv('images_human.csv', index=False, header=False)
````

````python
for x in df['deployment_id'].unique():
    Path(f'human_images/{x}').mkdir(exist_ok=True, parents=True)
````

````python
for x, y, z, t in zip(df['location'], df['deployment_id'], df['image_id'], df['timestamp']):
    if not y in [...]:
        continue
        
    out_path = f'human_images/{y}/{z}{Path(x).suffix}'
    shutil.copy2(f'human_images/{Path(x).name}', out_path)
    change_file_time(out_path, t)
````
